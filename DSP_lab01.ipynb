```python
import numpy as np
import matplotlib.pyplot as plt

def idft_matrix(N):
    """Generates the IDFT matrix of size NxN."""
    n = np.arange(N)
    k = n.reshape((N, 1))
    W = np.exp(2j * np.pi * k * n / N) / N
    return W

def synthesize_signal(x_mu):
    """Synthesizes a discrete-time signal using the IDFT matrix notation."""
    N = len(x_mu)
    W = idft_matrix(N)
    x_n = np.real(W @ x_mu)  #Compute the IDFT and take the real part
    return x_n, W

# Given signal
x_mu = np.array([6, 2, 4, 3, 4, 5, 0, 0, 0, 0], dtype=complex)

# Compute IDFT
x_n, W = synthesize_signal(x_mu)

#Plot the synthesized signal
plt.stem(range(len(x_n)), x_n)
plt.xlabel("n")
plt.ylabel("x[n]")
plt.title("Synthesized Signal using IDFT")
plt.grid()
plt.show()

#Display matrices W and K
N = len(x_mu)
k_values = np.arange(N).reshape((N, 1))  # Column vector
n_values = np.arange(N)  # Row vector

print("IDFT Matrix W:")
print(W)
print("\nIndex Matrix K:")
print(k_values)
```
