```python
import numpy as np
import matplotlib.pyplot as plt

# Parametry sygnau
f = 10  # czstotliwo sygnau [Hz]
fs = 15 # czstotliwo prbkowania [Hz]
T = 1   # czas trwania [s]
t = np.linspace(0, T, 1000)  # czas cigy
n = np.arange(0, T, 1/fs)     # chwile prbkowania

# Sygna oryginalny i sprbkowany
signal_original = np.sin(2 * np.pi * f * t)
signal_sampled = np.sin(2 * np.pi * f * n)

# Sygna z aliasem (po rekonstrukcji)
f_alias = abs(f - fs)  # aliasing daje 5 Hz
signal_alias = np.sin(2 * np.pi * f_alias * t)

# Wykresy
plt.figure(figsize=(10, 6))
plt.plot(t, signal_original, label="Oryginalny (10 Hz)", alpha=0.7)
plt.stem(n, signal_sampled, linefmt='r-', markerfmt='ro', basefmt=' ', label="Prbki (15 Hz)")
plt.plot(t, signal_alias, 'g--', label="Alias (5 Hz)")
plt.xlabel("Czas [s]")
plt.ylabel("Amplituda")
plt.title("Aliasing przy prbkowaniu 15 Hz dla sygnau 10 Hz")
plt.legend()
plt.grid()
plt.show()
```
